```python
import os
import geopandas as gpd
import pandas as pd
from osgeo import gdal
```

```python
# Define input and output file paths
input_vector = "/Users/klemenkubelj/Documents/school/graduate/masters/data/si_karte_globin/merged/globine_q10.gpkg"
output_raster = "/Users/klemenkubelj/Documents/school/graduate/masters/data/si_karte_globin/merged/OUTPUT.tif"

# Define rasterization options
options = gdal.RasterizeOptions(
    layers=["globine_q10"],
    attribute="k_globina",
    outputType=gdal.GDT_Float32,
    xRes=5.0,
    yRes=5.0,
    noData=0.0,
    outputBounds=[388122.3679, 42819.0, 599102.3819, 161472.723],
    format="GTiff"
)
```

```python

# Perform rasterization
gdal.Rasterize(output_raster, input_vector, options=options)
```
